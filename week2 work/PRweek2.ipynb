```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/week2 work/cardata.csv')
print("Dataset Preview:")
print(df.head())
```

```python
print("\nDataset Description:")
df.describe()
```

```python
#  Distribution of Selling Price
plt.figure(figsize=(8, 5))
plt.title('Selling Price Distribution')
sns.histplot(df['Selling_Price'], kde=True, color='teal')
plt.xlabel('Selling Price (in lakhs)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Relationship between Present Price and Selling Price
plt.figure(figsize=(8, 5))
plt.scatter(df['Present_Price'], df['Selling_Price'], color='darkorange')
plt.title('Present Price vs Selling Price')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.grid(True)
plt.show()
```

```python
X = df[['Present_Price']].values  # Feature
y = df['Selling_Price'].values   # Target
```

```python
# Train Linear Regression Model
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Train Polynomial Regression Model
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_poly, y)
```

```python
# Predictions
y_pred_lin = lin_reg.predict(X)
y_pred_poly = lin_reg_poly.predict(X_poly)
```

```python
# Visualize Linear Regression
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='lightcoral')
plt.plot(X, y_pred_lin, color='blue')
plt.title('Linear Regression: Selling vs Present Price')
plt.xlabel('Present Price')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
```

```python

# Visualize Polynomial Regression
X_grid = np.arange(min(X), max(X), 0.1).reshape(-1, 1)
X_grid_poly = poly.transform(X_grid)
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='lightcoral')
plt.plot(X_grid, lin_reg_poly.predict(X_grid_poly), color='green')
plt.title('Polynomial Regression (Degree 4)')
plt.xlabel('Present Price')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
```

```python
# Predict Selling Price for Present Price = 6.5
test_input = [[5]]
pred_lin = lin_reg.predict(test_input)
pred_poly = lin_reg_poly.predict(poly.transform(test_input))
print(f"ðŸ”¹ Linear Regression Prediction for Present Price 5: {pred_lin[0]:.2f} lakhs")
print(f"ðŸ”¹ Polynomial Regression Prediction for Present Price 5: {pred_poly[0]:.2f} lakhs")
```
