```python
import numpy as np
```

```python
# vector addtion
u= np.array([1,2])
v= np.array([3,4])
result=u+v
result
```

```python
# vector Multiplication
u= 3
v= np.array([2,-1])
result=u * v
result
```

```python
# dot Product
u= np.array([1,2])
v= np.array([3,4])
result=np.dot(u,v)
result
```

```python
# cross Product
u= np.array([1,0,0])
v= np.array([0,1,0])
result=np.cross(u,v)
result
```

```python
# norm of a vector
u= np.array([3,4])
result=np.linalg.norm(u)
result
```

```python
# Orthogonality condition
u= np.array([1,2])
v= np.array([-2,1])
result=np.dot(u,v)
is_orthogonal = result==0
is_orthogonal
```

```python
# Matrix Addition 2D array
u= np.array([[1,2],[3,4]])
v= np.array([[5,6],[7,8]])
result=u+v
result
```

```python
# Matrix Addition 2D array
u= 2
v= np.array([[5,6],[7,8]])
result=u * v
result
```

```python
# Matrix vector multiplication 2D array
A= np.array([[1,2],[3,4]])
x= np.array([5,6])
result=np.dot(A,x)
result
```

```python
# Matrix multiplication 2D array
A= np.array([[1,2],[3,4]])
x= np.array([[5,6],[7,8]])
result=np.dot(A,x)
result
```

```python
# Matrix transpose 2D array
A= np.array([[1,2],[3,4]])
result=A.T
result
```

```python
#determinant of 2 * 2 Matrix
A= np.array([[3,8],[4,6]])
result=np.linalg.det(A)
result
```
